```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv")
germany = df[df["country"] == "Germany"]
dataset = germany.tail(30)
```

```python
import matplotlib.pyplot as plt
year = dataset["year"]
co2 = dataset["co2"]
plt.figure(figsize=(10,6), dpi=100)
plt.plot(year, co2, color='royalblue', linewidth=3)

plt.title("Germany's Total CO₂ Emissions in the Last 30 Years")
plt.xlabel("Year")
plt.ylabel("CO₂ Emissions (million tonnes)")
```

```python
year = dataset["year"]
co2 = dataset["co2"]
population = dataset["population"]
plt.figure(figsize=(8,6))
plt.scatter(population/1e6, co2, color="blue")
plt.plot(population/1e6, co2, color="green")
plt.title("Germany: Population vs CO₂ Emissions")
plt.xlabel("Population (millions)")
plt.ylabel("CO₂ Emissions (million tonnes)")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
germany_last5 = df[df["country"] == "Germany"].tail(5)
uk_last5      = df[df["country"] == "United Kingdom"].tail(5)
plt.figure(figsize=(10,6))


plt.bar(germany_last5["year"], germany_last5["co2"], label="Germany", color="blue") # Germany bars


plt.bar(uk_last5["year"], uk_last5["co2"], label="United Kingdom", color="Black") # UK bars

plt.xlabel("Year")
plt.ylabel("CO₂ Emissions (million tonnes)")
plt.title("Germany vs UK CO₂ Emissions (Last 5 Years)")
plt.legend()

plt.show()

```
